```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/admin/Downloads/CLV_RFM_customer_orders.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['Order_date']=pd.to_datetime(data['Order_date'])
```

```python
from datetime import datetime
```

```python
rfm_cust=data.groupby('Customer').agg({'Order_date': lambda x: (datetime.today() - x.max()).days,'Order_id' : lambda x: len(x),
                                                              'Amount': lambda x: x.sum()})
```

```python
rfm_cust.head()
```

```python
data[data['Customer']=='Aakanksha']
```

```python
rfm_cust['Order_date']= rfm_cust['Order_date'].astype(int)
rfm_cust.columns=['recency','frquency','monetary_value']
rfm_cust.head()
```

```python
rfm_cust['R_quartile']=pd.qcut(rfm_cust['recency'],q=4,labels=['1','2','3','4'])
```

```python
rfm_cust.head()
```

```python
rfm_cust['M_quartile']=pd.qcut(rfm_cust['monetary_value'],q=4,labels=['4','3','2','1'])
```

```python
rfm_cust.head()
```

```python
rfm_cust['F_quartile']=pd.qcut(rfm_cust['frquency'].rank(method='first'),q=4,labels=['4','3','2','1'])
```

```python
rfm_cust.head()
```

```python
rfmlist=[]
for i in range(0,len(rfm_cust)):
    rfmlist.append(str(rfm_cust.iloc[i]['R_quartile'])+str(rfm_cust.iloc[i]['F_quartile'])+str(rfm_cust.iloc[i]['M_quartile']))
rfm_cust['RFM']=rfmlist 
```
