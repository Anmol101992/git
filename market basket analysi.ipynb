```python
data=[['Milk','Onion','Bread','Butter','Eggs','Yogurt'],
     ['Cheese','Onion','Bread','Butter','Eggs','Yogurt'],
     ['Milk','Apple','Butter','Eggs'],
     ['Milk','Orange','Corn','Butter','Yogurt'],
     ['Corn','Onion','Onion','Butter','Icecream','Eggs']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
!pip install mlxtend
```

```python
te= TransactionEncoder()
```

```python
datat=te.fit(data).transform(data)
```

```python
datat
```

```python
import pandas as pd
productdata=pd.DataFrame(datat,columns=te.columns_)
```

```python
productdata
```

```python
#Market basket analysis basically uses 3 different matrix measurnment -support,confident and lift
#support- it gives a fraction of records which contains factor,the factor can be either 1 or multiple,basically support
#tells us about frequent combition of factors.support equal to= no. of records or transactions containing factor (a or b )or both (a,b)/no. of total records
#confidence is equal to no of record containig both a & b/no.of records containing(a), it tells us about how often the factors a & b 
#occurs together given the no. of times a occurs.
#suppose a is bread and b is butter  the confidence will detect the no.of records when bread and butter together be purchased /no of times  when alone bread was purchased
# lift indicates the strength of a role over the random occurence of a & b, the greater the value of the lift , the more is the strength
# if lift is less than 1 there is no association if lift is equal to 1 this means that there is a very poor association 
# if lift is greater than 1 that means it is good association,if lift is equal to 3 it means that if you buy a,the chances of buying b is
# 3times, lift = support of x union y /support of x*support of y
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Determining the frequent factors
frequent_factors = apriori(productdata,use_colnames=True,min_support=0.1,max_len=4)
```

```python
#Filtering the factors on the basis of conditions
rules = association_rules(frequent_factors,metric ='lift',min_threshold=1)
```

```python
#Displaying the first 10 records
print(rules.head(10).sort_values(by='confidence',ascending =False))
```

```python
#lift of cheese & bread:
# support of(cheese union bread)/support of cheese* support of bread
# 0.2/(0.2*0.4)=2.5
```
